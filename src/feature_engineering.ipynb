```python
# src/feature_engineering.py

import pandas as pd
from sklearn.preprocessing import LabelEncoder

def load_data(filepath):
    """
    Load the dataset into a pandas DataFrame.
    """
    df = pd.read_csv(filepath)
    return df

def handle_missing_values(df):
    """
    Handle missing values appropriately.
    """
    df_col=df.columns
    df_col_num= []

    for col in df_col:
        if df[col].dtype !='object':
            df_col_num.append(col)
        else:
            continue
    

    df_num_2=df[df_col_num]
    df[df_col_num].fillna(df[df_col_num].mean(),inplace=True)
    
    
    return df

def create_age_groups(df):
    bins = [17, 30, 45, 60, 100]  # Age ranges
    labels = ['young', 'adult', 'middle-aged', 'senior']  # Labels for age groups
    #todo: create age groups
    df['age_groups']=pd.cut(df['age'],bins=bins,labels=labels)

    df = df.dropna()
    return df

def encode_categorical_features(df):
    """
    Encode categorical features using Label Encoding.
    """
    le = LabelEncoder()
    categorical_cols = ['country', 'gender', 'credit_card','age_groups']
    df[categorical_cols]=le.fit_transform(categorical_cols)
   
    # todo: run a loop to fit each column into le
    return df

def save_processed_data(df, filepath):
    """
    Save the processed DataFrame to a CSV file.
    """

    df.to_csv(filepath)
    #todo save the processed data into data folder using to_csv

def main():
    # Load data
    df = load_data('/workspaces/ts-aiml-datascience-challenge-Shantanu467/data/bank_churn.csv')

    
    # Handle missing values
    df = handle_missing_values(df)
    
    # Create age groups
    df = create_age_groups(df)
    
    # Encode categorical features
    df = encode_categorical_features(df)
    print(df.head(10))
    
    # Save processed data
    save_processed_data(df, '/workspaces/ts-aiml-datascience-challenge-Shantanu467/data/processed_bank_churn.csv') 

if __name__ == "__main__":
    main()
```
